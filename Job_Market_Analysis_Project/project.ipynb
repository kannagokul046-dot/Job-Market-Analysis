```python
import pandas as pd 
import numpy as np 
```

```python
#uploding data into a python
job_data = pd.read_csv(r"E:\NaukriData_Data Science.csv")

#find a rows and columns 
job_data.shape

```

```python
# top 10 rows 
job_data.head(10)
# column names 
job_data.columns
```

```python
# getting a info of a data 
job_data.info()
```

```python
#Rename columns
job_data.columns = (
    job_data.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
)

job_data.columns
```

```python
#Handle missing values
job_data.isna().sum()

job_data.replace(
    ["Not disclosed", "None", "none", "null", ""],
    np.nan,
    inplace=True
)

job_data 
```

```python
# rechecked
job_data.isna().sum()
```

```python
#Clean EXPERIENCE column
job_data["experience"] = (
    job_data["experience_required"]
    .str.extract(r"(\d+)")
    .astype(float)
)
job_data["experience"].head()
```

```python
#clean loctation column 

job_data["locations"] = (
    job_data["locations"]
    .str.split("/")
    .str[0]
)

job_data["locations"].head(2)
```
